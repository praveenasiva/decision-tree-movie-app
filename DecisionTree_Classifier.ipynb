```python
import pandas as pd
movie_info = pd.read_csv('/content/Movie_Interests_DecisionTree.csv')
movie_info
```

```python
input_dataset = movie_info.drop(columns=['Interest'])
```

```python
input_dataset
```

```python
output_dataset = movie_info['Interest']
output_dataset
```

```python
from sklearn.tree import DecisionTreeClassifier
movie_model = DecisionTreeClassifier()
movie_model.fit(input_dataset,output_dataset)
movie_interest = movie_model.predict([[14,1],[33,0]])
movie_interest
```
